3.WAP to detect single space in the string.

```python
def sap(name):
    for i in range(len(name)):
        if name[i].isspace():
            return True
```

```python
sap("cdnancnd")
```

4.WAP to detect double space in the string

```python
def saps(name):
    if "  " in name:
        return True
    else:
        return False
```

```python
saps("apurva  a")
```
